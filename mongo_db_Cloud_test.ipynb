## 3. movi_database
-'./app/database/movies_database.pkl'

```python
def get_documents(path):
    with open(path, 'rb') as file:
        data = pickle.load(file)
    documents = [{key: values} for key, values in data.items()]
    return documents
```

```python
path = './app/database/movies_database.pkl'
movie_documents = get_documents(path)
```

```python
def insert_documents(collection, documents):
    x = collection.insert_many(documents)
    print('ids of inserted documents\n---------------------')
    for id in x.inserted_ids:
        print(id)
```

```python
movie_collection = client['final_project']['movie']
insert_documents(movie_collection, movie_documents)
```

```python
def get_databse(collection):
    database = defaultdict(list)
    for i in collection.find():
        key = list(i.keys())[1]
        value = i[key]
        database[key] = value
    return database
```

```python
movie_collection = client['final_project']['movie']
movies_database = get_databse(movie_collection)
```
