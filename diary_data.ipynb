```python
label_id_lst = []
for index, text in enumerate(diary_lst):
    print(f'text : \n {text}')
    labels = list(input().split(','))
    labels_id = [labels_to_id[key] for key in labels]
    label_id_lst.append(labels_id)
    print(f'labels : {labels_id}')
```

```python
with open('./label_id_lst_temp.pickle', 'wb') as f:
    pickle.dump(label_id_lst, f)
```

```python
df = pd.DataFrame((zip(diary_lst, label_id_lst)), columns = ['text', 'labels'])
print(df)
```
