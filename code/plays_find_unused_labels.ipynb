```python
import pandas as pd
import os
```

```python
dir_path = "../data/plays"
dataset_list = os.listdir(dir_path)
dataset_list
```

```python
dataset = pd.read_csv(f"{dir_path}/{dataset_list[0]}")
for dataset_name in dataset_list[1:]:
    temp_dataset = pd.read_csv(f"{dir_path}/{dataset_name}")
    dataset = pd.concat([dataset, temp_dataset])

dataset
```

```python
# 중복 제거
dataset = dataset.drop_duplicates(ignore_index=True)
dataset
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline

model_name = "searle-j/kote_for_easygoing_people"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

pipe = TextClassificationPipeline(
        model=model,
        tokenizer=tokenizer,
        device=0, # gpu number, -1 if cpu used
        return_all_scores=True,
        # function_to_apply='sigmoid'
    )
```

```python
LABELS = ['불평/불만',
 '환영/호의',
 '감동/감탄',
 '지긋지긋',
 '고마움',
 '슬픔',
 '화남/분노',
 '존경',
 '기대감',
 '우쭐댐/무시함',
 '안타까움/실망',
 '비장함',
 '의심/불신',
 '뿌듯함',
 '편안/쾌적',
 '신기함/관심',
 '아껴주는',
 '부끄러움',
 '공포/무서움',
 '절망',
 '한심함',
 '역겨움/징그러움',
 '짜증',
 '어이없음',
 '없음',
 '패배/자기혐오',
 '귀찮음',
 '힘듦/지침',
 '즐거움/신남',
 '깨달음',
 '죄책감',
 '증오/혐오',
 '흐뭇함(귀여움/예쁨)',
 '당황/난처',
 '경악',
 '부담/안_내킴',
 '서러움',
 '재미없음',
 '불쌍함/연민',
 '놀람',
 '행복',
 '불안/걱정',
 '기쁨',
 '안심/신뢰']

plays_by_label = {k: [] for k in LABELS}
plays_by_label
```

```python
for _, play in dataset.iterrows():
    # print(play)
    title = play['title']
    sub_title = play['sub_title']
    intro = play['intro']

    outputs = sorted(pipe(str(intro)[:512])[0], key=lambda x: x['score'], reverse=True)
    for output in outputs[:5]: # index를 조절하면 top-k를 조절할 수 있습니다.
        plays_by_label[output['label']].append([title, sub_title])

plays_by_label
```

```python
no_play_emotion = []
under_22_play_emotion = []
for emotion in plays_by_label.keys():
    if len(plays_by_label[emotion]) == 0:
        no_play_emotion.append(emotion)
    elif len(plays_by_label[emotion]) < 13:
        under_22_play_emotion.append(emotion)

print(f"어떤 공연도 포함되지 않은 감정:\n\t{no_play_emotion}")
print(f"22개 미만(934개 공연/44개 감정=21.xxxx)의 공연이 포함된 감정:\n\t{under_22_play_emotion}")
    
```
