```python
driver = webdriver.Safari()
```

```python
from time import sleep

base_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220526'  # 맨 뒤 data 변수를 원하는 날짜로 바꿔줘야 합니다.
driver.get(base_url)

movies = driver.find_elements_by_class_name('tit5')
links = list(map(lambda x: x.find_element_by_tag_name('a').get_attribute('href'), movies))

rec_movies = []
for link in links:
    driver.get(link)
    title = driver.find_element_by_class_name('h_movie').find_element_by_tag_name('a').text
    text = driver.find_element_by_class_name('con_tx').text
    text = text.replace("\n", " ")
    text = text.replace("\xa0", " ")
    text = text.replace("  ", " ")
    rec_movies.append((title, text, link))

rec_movies
```

```python
len(rec_movies)
```

```python
rec_movies = [{'title': movie[0], 'text': movie[1], 'link_url': movie[2]} for movie in rec_movies]
rec_movies
    
```

```python
import pandas as pd

df = pd.DataFrame(rec_movies)
df
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df)
dataset.save_to_disk("top40_movies_by_score")
```
