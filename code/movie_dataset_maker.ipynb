```python
import pandas as pd
from datasets import load_from_disk
```

```python
dataset1 = load_from_disk("../data/top40_movies_by_score")
dataset1 = dataset1.to_pandas()
dataset1
```

```python
dataset2 = pd.read_csv("../data/masterpieces.csv")
dataset2
```

```python
dataset3 = pd.read_csv("../data/popular_search.csv")
dataset3.rename(columns = {'intro': 'text'}, inplace=True)
dataset3
```

```python
dataset = pd.concat([dataset1, dataset2, dataset3]).drop_duplicates(ignore_index=True)
dataset
```

```python
dataset.to_csv("movie_dataset.csv", index=False)
```
