```python
import pandas as pd
import os
```

```python
dir_path = "/opt/ml/final-project/final-project-level3-nlp-01/data/music_data"
song_file_list = os.listdir(dir_path)
song_file_list
```

```python
first_song_file = song_file_list[0]
song_df = pd.read_csv(f"{dir_path}/{first_song_file}")
song_df
```

```python

for song_file in song_file_list[1:]:
    temp_df = pd.read_csv(f"{dir_path}/{song_file}")
    song_df = pd.concat([song_df, temp_df], ignore_index=True)
    
song_df
```

```python
song_df = song_df.drop_duplicates(ignore_index=True)
song_df
```

```python
song_df.to_csv("../data/songs/song_dataset_image.csv", index=False)
```
